### Concat & Merging

```python
# This is how to concat two datasets with the help of pandas



df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                    index=[0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                    index=[4,5,6,7])
# frames = [df1,df2]
result = pd.concat([df1,df2])
```

```python
df1
```

```python
result
```

```python
'''


table1
cust_id, cust_name
1, satyajit
2, hillary
3, raj
4, shankar

table2
cust_id, salary
2, 10000
3, 25000


Output:- 
cust_id, cust_name, salary
if i do a inner joint 

2, hillary, 10000
3, Raj      25000

if i do a left joint
1, satyjit, -
2, hillary, 10000
3, raj, 25000
4, shankar, -

if i do a right join
2, hillary, 10000
3, raj, 25000


'''

merge_df = pd.merge(df1, df2, on='C',how='left')
merge_df
```

## loc vs iloc 

The main distinction between the two methods are :

* loc gets rows(and/or columns) with particular labels
* iloc gets rows (and/or columns) at integer locations.

```python
s = pd.Series(list("abcdef"), index = [49,48,47,0,1,2])
s

# s.loc[0]  # value at index label 0,    output = d
# s.iloc[0]   # value at index location 0     output = a

# s.iloc[3::]
s.iloc[::3]
```
